# Select variables suitable for exploratory visual analysis
For this exploratory analysis, we will focus on the following variables:

Year

Month

COVID-19 Deaths

Number of Mentions

State

Condition Group

Age Group

We will discard other columns that are not needed for this analysis, such as Data As Of, Start Date, End Date, and ICD10_codes.

```python
# Select relevant columns for analysis
df_covid_analysis = df_covid_final_cleaned[['Year', 'Month', 'COVID-19 Deaths', 'Number of Mentions', 'State', 'Condition Group', 'Age Group']]

# Display the first few rows of the selected dataframe
df_covid_analysis.head()
```

# Creating a correlation matrix heatmap

```python
# Create a correlation matrix
corr_matrix = df_covid_analysis[['Year', 'Month', 'COVID-19 Deaths','Number of Mentions']].corr()

# Plot the correlation matrix heatmap
plt.figure(figsize=(7, 5))
sns.heatmap(corr_matrix, annot=True, cmap='magma', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

The correlation matrix shows the pairwise correlations between numerical variables in the dataset. The coefficients range from -1 to 1, where:

1 indicates a perfect positive correlation.
-1 indicates a perfect negative correlation.
0 indicates no correlation.
From the heatmap, we can see that:

COVID-19 Deaths and Number of Mentions have a strong positive correlation.
Year and Month have no significant correlation with COVID-19 Deaths or Number of Mentions.

# Creating a scatterplot for variables with the strongest correlations

```python
# Scatterplot for COVID-19 Deaths and Number of Mentions
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_covid_analysis, x='Number of Mentions', y='COVID-19 Deaths')
plt.title('Scatterplot of COVID-19 Deaths vs Number of Mentions')
plt.xlabel('Number of Mentions')
plt.ylabel('COVID-19 Deaths')
plt.show()
```

The scatterplot shows a positive relationship between COVID-19 Deaths and Number of Mentions. This indicates that as the number of mentions of a condition increases, the number of COVID-19 deaths also tends to increase. The relationship appears to be linear with some dispersion.

# Create a pair plot of the entire data set

```python
# Create a pair plot
sns.pairplot(df_covid_analysis[['Year', 'Month', 'COVID-19 Deaths', 'Number of Mentions']])
plt.show()
```

The pair plot provides a visual summary of the distribution of each numerical variable and the relationships between them. Well, it is hard to find the relation between different variables. The diagonal plots show the distribution of each variable, while the off-diagonal plots show the relationships between pairs of variables. The COVID-19 Deaths and Number of Mentions relationship stands out as having a clear positive correlation. The distributions of Year and Month are more uniform and do not show strong relationships with other variables.

 # Creating a categorical plot and interpret the results

```python
# Categorical plot for COVID-19 Deaths by Age Group
plt.figure(figsize=(12, 6))
sns.boxplot(data=df_covid_analysis, x='Age Group', y='COVID-19 Deaths')
plt.title('Boxplot of COVID-19 Deaths by Age Group')
plt.xlabel('Age Group')
plt.ylabel('COVID-19 Deaths')
plt.xticks(rotation=45)
plt.show()
```

The boxplot shows the distribution of COVID-19 Deaths across different age groups. The median number of deaths increases with age, with the highest number of deaths occurring in the oldest age groups. There are also some outliers, particularly in the older age groups, indicating variability in the number of deaths within those groups.
